```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

import plotly.express as px # Importando plotly para o treemap



sns.set(style="whitegrid")

df = pd.read_csv("MODULO7_PROJETOFINAL_BASE_SUPERMERCADO - MODULO7_PROJETOFINAL_BASE_SUPERMERCADO (1).csv.csv")



# --- EXERCÍCIO 1: Média e Mediana por Categoria ---

resumo_media = df.groupby('Categoria')['Preco_Normal'].mean().reset_index()

resumo_media.columns = ['Categoria', 'Media_Preco_Normal']



resumo_mediana = df.groupby('Categoria')['Preco_Normal'].median().reset_index()

resumo_mediana.columns = ['Categoria', 'Mediana_Preco_Normal']

# Unir as duas tabelas

resumo = pd.merge(resumo_media, resumo_mediana, on='Categoria')



print("=== Média e Mediana do Preço Normal por Categoria ===")

print(resumo)



# Identificar categorias com média acima da mediana

categorias_acima = resumo[resumo['Media_Preco_Normal'] > resumo['Mediana_Preco_Normal']]

print("\nCategorias com média acima da mediana:")

print(categorias_acima['Categoria'].tolist())



# --- EXERCÍCIO 2: Desvio Padrão + Boxplot + Outliers ---



# Calcular desvio padrão do Preco_Normal por categoria

desvio_preco = df.groupby('Categoria')['Preco_Normal'].std().reset_index()

desvio_preco.columns = ['Categoria', 'Desvio_Padrao_Preco_Normal']



# Mostrar categoria com maior variação

maior_variacao = desvio_preco.sort_values(by='Desvio_Padrao_Preco_Normal', ascending=False)

print("\n=== Desvio Padrão por Categoria ===")

print(desvio_preco)



print("\nCategoria com maior variação de preço:")

print(maior_variacao.head(1))



# Boxplot para visualizar a distribuição e outliers (todas as categorias)

plt.figure(figsize=(12, 6))

sns.boxplot(data=df, x='Categoria', y='Preco_Normal')

plt.title("Distribuição dos Preços Normais por Categoria (Boxplot Geral)")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()



# Boxplot específico da categoria com maior desvio padrão

# Pega a primeira categoria após a ordenação decrescente do desvio padrão

categoria_max_dp = maior_variacao.iloc[0]['Categoria']

df_categoria_max_dp = df[df['Categoria'] == categoria_max_dp]



plt.figure(figsize=(8, 6))

sns.boxplot(data=df_categoria_max_dp, y='Preco_Normal')

plt.title(f'Distribuição do Preço Normal - {categoria_max_dp} (Maior Variação)')

plt.ylabel('Preço Normal')

plt.tight_layout()

plt.show()



# Comentário sobre os outliers

print(f"\nNo boxplot da categoria '{categoria_max_dp}', os pontos que aparecem acima ou abaixo das hastes (whiskers) podem ser considerados outliers, indicando preços significativamente diferentes da maioria dos produtos nessa categoria.")


# --- EXERCÍCIO: Média de Desconto por Categoria (Gráfico de Barras) ---



# Calcular média de desconto por categoria

media_descontos = df.groupby('Categoria')['Desconto'].mean().reset_index()



# Gráfico de barras para a média de desconto

plt.figure(figsize=(10, 6))

sns.barplot(data=media_descontos, x='Categoria', y='Desconto')

plt.title("Média de Desconto por Categoria")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()


# --- EXERCÍCIO: Total de Desconto por Categoria (Treemap) ---

pio.renderers.default = 'iframe_connected'

descontos_categoria = df.groupby(['Categoria', 'Marca'])['Desconto'].mean().reset_index()

# Filtrando os valores de desconto que são iguais a zero
descontos_categoria = descontos_categoria[descontos_categoria['Desconto'] > 0]

# Criando o gráfico de treemap
fig = px.treemap(descontos_categoria,
         path=['Categoria', 'Marca'],
         values='Desconto',
         title="Média de Desconto por Categoria e Marca",
         color='Desconto',
         color_continuous_scale='Viridis')

# Exibindo o gráfico
fig.show()




```
